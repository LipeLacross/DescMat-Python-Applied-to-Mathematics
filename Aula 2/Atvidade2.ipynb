```python
%%writefile calculadora.py

import math

def calculadora(operacao, op1=0, op2=0, op3=0, op4=0):
    if operacao == 'soma':
        return soma(op1, op2)
    elif operacao == 'subtracao':
        return subtracao(op1, op2)
    elif operacao == 'multiplicacao':
        return multiplicacao(op1, op2)
    elif operacao == 'divisao':
        return divisao(op1, op2)
    elif operacao == 'potenciacao':
        return potenciacao(op1, op2)
    elif operacao == 'radiciacao':
        return radiciacao(op1, op2)
    elif operacao == 'exponencial':
        return exponencial(op1)
    elif operacao == 'modulo':
        return modulo(op1, op2)
    elif operacao == 'divisao_inteira':
        return divisao_inteira(op1, op2)
    elif operacao == 'conversao_graus_radianos':
        return conversao_graus_radianos(op1)
    elif operacao == 'conversao_radianos_graus':
        return conversao_radianos_graus(op1)
    elif operacao == 'calculo_hipotenusa':
        return calculo_hipotenusa(op1, op2)
    elif operacao == 'arredondamento':
        return arredondamento(op1)
    elif operacao == 'arredondamento_para_cima':
        return arredondamento_para_cima(op1)
    elif operacao == 'arredondamento_para_baixo':
        return arredondamento_para_baixo(op1)
    elif operacao == 'maior':
        return maior(op1, op2)
    elif operacao == 'menor':
        return menor(op1, op2)
    elif operacao == 'distancia_entre_pontos':
        return distancia_entre_pontos(op1, op2, op3, op4)
    elif operacao == 'determinante_matriz_2x2':
        return determinante_matriz_2x2(op1, op2, op3, op4)
    elif operacao == 'solucao_equacao_primeiro_grau':
        return solucao_equacao_primeiro_grau(op1, op2, op3)
    elif operacao == 'area_retangulo':
        return area_retangulo(op1, op2)
    elif operacao == 'perimetro_retangulo':
        return perimetro_retangulo(op1, op2)
    elif operacao == 'area_circulo':
        return area_circulo(op1)
    elif operacao == 'circunferencia_circulo':
        return circunferencia_circulo(op1)
    elif operacao == 'verificacao_triangulo':
        return verificacao_triangulo(op1, op2, op3)
    elif operacao == 'verificacao_numeros_primos':
        return verificacao_numeros_primos(op1)
    elif operacao == 'conversor_unidades':
        return conversor_unidades(op1, op2)
    elif operacao == 'juros_simples':
        return juros_simples(op1, op2, op3)
    elif operacao == 'juros_compostos':
        return juros_compostos(op1, op2, op3)
    elif operacao == 'soma_numeros_naturais':
        return soma_numeros_naturais(op1)
    elif operacao == 'fatorial':
        return fatorial(op1)
    elif operacao == 'permutacao':
        return permutacao(op1, op2)
    elif operacao == 'combinacao':
        return combinacao(op1, op2)
    elif operacao == 'conversor_bases':
        return conversor_bases(op1, op2)
    elif operacao == 'media':
        return media(op1)
    elif operacao == 'desvio_padrao':
        return desvio_padrao(op1)
    elif operacao == 'tangente_reta':
        return tangente_reta(op1, op2, op3)
    else:
        raise ValueError("Operação não suportada")

# Operações básicas
def soma(o1, o2):
    return o1 + o2

def subtracao(o1, o2):
    return o1 - o2

def multiplicacao(o1, o2):
    return o1 * o2

def divisao(o1, o2):
    if o2 == 0:
        raise ValueError("Divisão por zero")
    return o1 / o2

def potenciacao(base, exp):
    return base ** exp

def radiciacao(num, raiz):
    if num < 0:
        raise ValueError("Não é possível calcular a raiz de um número negativo")
    return num ** (1 / raiz)

def exponencial(x):
    return math.exp(x)

def modulo(a, b):
    return a % b

def divisao_inteira(a, b):
    return a // b

def conversao_graus_radianos(graus):
    return math.radians(graus)

def conversao_radianos_graus(radianos):
    return math.degrees(radianos)

def calculo_hipotenusa(a, b):
    return math.sqrt(a ** 2 + b ** 2)

def arredondamento(valor):
    return round(valor)

def arredondamento_para_cima(valor):
    return math.ceil(valor)

def arredondamento_para_baixo(valor):
    return math.floor(valor)

def maior(a, b):
    return max(a, b)

def menor(a, b):
    return min(a, b)

def distancia_entre_pontos(x1, y1, x2, y2):
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def determinante_matriz_2x2(a, b, c, d):
    return a * d - b * c

def solucao_equacao_primeiro_grau(a, b, c):
    if a == 0:
        raise ValueError("Coeficiente 'a' não pode ser zero")
    return (c - b) / a

# Operações geométricas
def area_retangulo(largura, altura):
    return largura * altura

def perimetro_retangulo(largura, altura):
    return 2 * (largura + altura)

def area_circulo(raio):
    return math.pi * raio ** 2

def circunferencia_circulo(raio):
    return 2 * math.pi * raio

def verificacao_triangulo(a, b, c):
    return a + b > c and a + c > b and b + c > a

# Operações numéricas e financeiras
def verificacao_numeros_primos(num):
    if num <= 1:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

def conversor_unidades(valor, tipo):
    if tipo == 'milhas_para_km':
        return valor * 1.60934
    elif tipo == 'metros_para_pes':
        return valor * 3.28084
    else:
        raise ValueError("Conversão não suportada")

def juros_simples(principal, taxa, tempo):
    return principal * taxa * tempo

def juros_compostos(principal, taxa, tempo):
    return principal * (1 + taxa) ** tempo

def soma_numeros_naturais(n):
    if n <= 0:
        return 0
    return n * (n + 1) // 2

def fatorial(n):
    if n == 0:
        return 1
    return n * fatorial(n - 1)

def permutacao(n, r):
    return fatorial(n) // fatorial(n - r)

def combinacao(n, r):
    return fatorial(n) // (fatorial(r) * fatorial(n - r))

def conversor_bases(num, base):
    if base == 'decimal_para_binario':
        return bin(num)[2:]
    elif base == 'decimal_para_hexadecimal':
        return hex(num)[2:].upper()
    else:
        raise ValueError("Conversão de base não suportada")

def media(valores):
    return sum(valores) / len(valores)

def desvio_padrao(valores):
    media_valores = media(valores)
    variancia = sum((x - media_valores) ** 2 for x in valores) / len(valores)
    return math.sqrt(variancia)

def tangente_reta(a, b, c):
    return -a / b
```
