```python
%%writefile calculadora.py

import math

def calculadora(operacao, *args):
    if operacao == 'soma':
        return soma(*args)
    elif operacao == 'subtracao':
        return subtracao(*args)
    elif operacao == 'multiplicacao':
        return multiplicacao(*args)
    elif operacao == 'divisao':
        if args[1] == 0:
            raise ValueError("Divisão por zero não é permitida.")
        return divisao(*args)
    elif operacao == 'potenciacao':
        return potenciacao(*args)
    elif operacao == 'radiciacao':
        if args[0] < 0 and args[1] % 2 == 0:
            raise ValueError("Não é possível calcular a raiz de um número negativo para índices pares.")
        return radiciacao(*args)
    elif operacao == 'exponencial':
        return exponencial(args[0])
    elif operacao == 'modulo':
        return modulo(*args)
    elif operacao == 'divisao_inteira':
        return divisao_inteira(*args)
    elif operacao == 'conversao_graus_radianos':
        return conversao_graus_radianos(args[0])
    elif operacao == 'conversao_radianos_graus':
        return conversao_radianos_graus(args[0])
    elif operacao == 'calculo_hipotenusa':
        return calculo_hipotenusa(*args)
    elif operacao == 'arredondamento':
        return arredondamento(args[0])
    elif operacao == 'arredondamento_para_cima':
        return arredondamento_para_cima(args[0])
    elif operacao == 'arredondamento_para_baixo':
        return arredondamento_para_baixo(args[0])
    elif operacao == 'maior':
        return maior(*args)
    elif operacao == 'menor':
        return menor(*args)
    elif operacao == 'distancia_entre_pontos':
        return distancia_entre_pontos(*args)
    elif operacao == 'determinante_matriz_2x2':
        return determinante_matriz_2x2(*args)
    elif operacao == 'solucao_equacao_primeiro_grau':
        return solucao_equacao_primeiro_grau(*args)
    elif operacao == 'area_retangulo':
        return area_retangulo(*args)
    elif operacao == 'perimetro_retangulo':
        return perimetro_retangulo(*args)
    elif operacao == 'area_circulo':
        return area_circulo(args[0])
    elif operacao == 'circunferencia_circulo':
        return circunferencia_circulo(args[0])
    elif operacao == 'verificacao_triangulo':
        return verificacao_triangulo(*args)
    elif operacao == 'verificacao_numeros_primos':
        return verificacao_numeros_primos(args[0])
    elif operacao == 'conversor_unidades':
        return conversor_unidades(args[0], args[1])
    elif operacao == 'juros_simples':
        return juros_simples(*args)
    elif operacao == 'juros_compostos':
        return round(juros_compostos(*args), 2)  # Ajuste de precisão
    elif operacao == 'soma_numeros_naturais':
        return soma_numeros_naturais(args[0])
    elif operacao == 'fatorial':
        return fatorial(args[0])
    elif operacao == 'permutacao':
        return permutacao(*args)
    elif operacao == 'combinacao':
        return combinacao(*args)
    elif operacao == 'conversor_bases':
        return conversor_bases(args[0], args[1])
    elif operacao == 'media':
        return media(*args)  # Suporta número variável de argumentos
    elif operacao == 'desvio_padrao':
        return desvio_padrao(*args)  # Suporta número variável de argumentos
    elif operacao == 'tangente_reta':
        return tangente_reta(*args)
    else:
        raise ValueError("Operação não reconhecida.")


def soma(*args):
    return sum(args)

def subtracao(o1, o2):
    return o1 - o2

def multiplicacao(*args):
    result = 1
    for num in args:
        result *= num
    return result

def divisao(o1, o2):
    return o1 / o2

def potenciacao(o1, o2):
    return o1 ** o2

def radiciacao(o1, o2):
    return o1 ** (1/o2)

def exponencial(o1):
    return math.exp(o1)

def modulo(o1, o2):
    return o1 % o2

def divisao_inteira(o1, o2):
    return o1 // o2

def conversao_graus_radianos(o1):
    return math.radians(o1)

def conversao_radianos_graus(o1):
    return math.degrees(o1)

def calculo_hipotenusa(o1, o2):
    return math.hypot(o1, o2)

def arredondamento(o1):
    return round(o1)

def arredondamento_para_cima(o1):
    return math.ceil(o1)

def arredondamento_para_baixo(o1):
    return math.floor(o1)

def maior(*args):
    return max(args)

def menor(*args):
    return min(args)

def distancia_entre_pontos(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def determinante_matriz_2x2(a, b, c, d):
    return a * d - b * c

def solucao_equacao_primeiro_grau(a, b, c):
    if a == 0:
        raise ValueError("Coeficiente 'a' não pode ser zero.")
    return (c - b) / a

def area_retangulo(largura, altura):
    return largura * altura

def perimetro_retangulo(largura, altura):
    return 2 * (largura + altura)

def area_circulo(raio):
    return math.pi * raio ** 2

def circunferencia_circulo(raio):
    return 2 * math.pi * raio

def verificacao_triangulo(a, b, c):
    return a + b > c and a + c > b and b + c > a

def verificacao_numeros_primos(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def conversor_unidades(valor, tipo):
    conversoes = {
        'milhas_para_km': valor * 1.60934,
        'metros_para_pes': valor * 3.28084,
    }
    return conversoes.get(tipo, None)

def juros_simples(principal, taxa, tempo):
    return principal * taxa * tempo

def juros_compostos(principal, taxa, tempo):
    return principal * (1 + taxa) ** tempo

def soma_numeros_naturais(n):
    if n <= 1:
        return n
    return n + soma_numeros_naturais(n - 1)

def fatorial(n):
    if n == 0:
        return 1
    return n * fatorial(n - 1)

def permutacao(n, r):
    return fatorial(n) // fatorial(n - r)

def combinacao(n, r):
    return fatorial(n) // (fatorial(r) * fatorial(n - r))

def conversor_bases(numero, tipo):
    if tipo == 'decimal_para_binario':
        return bin(numero)[2:]
    elif tipo == 'decimal_para_hexadecimal':
        return hex(numero)[2:].upper()
    return None

def media(*numeros):
    return sum(numeros) / len(numeros)

def desvio_padrao(*numeros):
    media_val = media(*numeros)
    return math.sqrt(sum((x - media_val) ** 2 for x in numeros) / len(numeros))

def tangente_reta(a, b, c):
    return -a / b if b != 0 else float('inf')


```

```python
%%writefile test_script.py
import pytest
from calculadora import calculadora
import math

# Operações Básicas
def test_calculadora_soma():
    assert calculadora('soma', 2, 3) == 5
    assert calculadora('soma', -1, 1) == 0

def test_calculadora_subtracao():
    assert calculadora('subtracao', 5, 3) == 2
    assert calculadora('subtracao', 0, 5) == -5

def test_calculadora_multiplicacao():
    assert calculadora('multiplicacao', 4, 3) == 12
    assert calculadora('multiplicacao', -1, 5) == -5

def test_calculadora_divisao():
    assert calculadora('divisao', 10, 2) == 5
    assert calculadora('divisao', 5, 2) == 2.5
    with pytest.raises(ValueError):
        calculadora('divisao', 5, 0)

def test_calculadora_potenciacao():
    assert calculadora('potenciacao', 2, 3) == 8
    assert calculadora('potenciacao', 4, 0.5) == 2

# Radiciação
def test_calculadora_radiciacao():
    assert calculadora('radiciacao', 16, 2) == 4
    assert calculadora('radiciacao', 27, 3) == 3
    with pytest.raises(ValueError):
        calculadora('radiciacao', -1, 2)

def test_calculadora_exponencial():
    assert calculadora('exponencial', 1) == pytest.approx(math.e)
    assert calculadora('exponencial', 0) == 1

# Operações Diversas
def test_calculadora_modulo():
    assert calculadora('modulo', 10, 3) == 1
    assert calculadora('modulo', -10, 3) == 2

def test_calculadora_divisao_inteira():
    assert calculadora('divisao_inteira', 10, 3) == 3
    assert calculadora('divisao_inteira', -10, 3) == -4

def test_calculadora_conversao_graus_radianos():
    assert calculadora('conversao_graus_radianos', 180) == math.pi
    assert calculadora('conversao_graus_radianos', 90) == math.pi / 2

def test_calculadora_conversao_radianos_graus():
    assert calculadora('conversao_radianos_graus', math.pi) == 180
    assert calculadora('conversao_radianos_graus', math.pi / 2) == 90

def test_calculadora_calculo_hipotenusa():
    assert calculadora('calculo_hipotenusa', 3, 4) == 5
    assert calculadora('calculo_hipotenusa', 5, 12) == 13

def test_calculadora_arredondamento():
    assert calculadora('arredondamento', 3.14) == 3
    assert calculadora('arredondamento', 3.5) == 4

def test_calculadora_arredondamento_para_cima():
    assert calculadora('arredondamento_para_cima', 3.14) == 4
    assert calculadora('arredondamento_para_cima', -3.14) == -3

def test_calculadora_arredondamento_para_baixo():
    assert calculadora('arredondamento_para_baixo', 3.14) == 3
    assert calculadora('arredondamento_para_baixo', -3.14) == -4

def test_calculadora_maior_menor():
    assert calculadora('maior', 2, 3) == 3
    assert calculadora('menor', 2, 3) == 2

def test_calculadora_distancia_entre_pontos():
    assert calculadora('distancia_entre_pontos', 0, 0, 3, 4) == 5
    assert calculadora('distancia_entre_pontos', 1, 1, 4, 5) == 5

def test_calculadora_determinante_matriz_2x2():
    assert calculadora('determinante_matriz_2x2', 1, 2, 3, 4) == -2
    assert calculadora('determinante_matriz_2x2', 5, 6, 7, 8) == -2

def test_calculadora_solucao_equacao_primeiro_grau():
    assert calculadora('solucao_equacao_primeiro_grau', 2, 3, 4) == 0.5
    assert calculadora('solucao_equacao_primeiro_grau', 1, -2, 3) == 5

# Operações de Cálculo Geométrico
def test_calculadora_area_retangulo():
    assert calculadora('area_retangulo', 5, 3) == 15
    assert calculadora('area_retangulo', 7, 2) == 14

def test_calculadora_perimetro_retangulo():
    assert calculadora('perimetro_retangulo', 5, 3) == 16
    assert calculadora('perimetro_retangulo', 7, 2) == 18

def test_calculadora_area_circulo():
    assert calculadora('area_circulo', 3) == pytest.approx(28.274333882308138, rel=1e-9)
    assert calculadora('area_circulo', 1) == pytest.approx(3.141592653589793, rel=1e-9)

def test_calculadora_circunferencia_circulo():
    assert calculadora('circunferencia_circulo', 3) == pytest.approx(18.84955592153876, rel=1e-9)
    assert calculadora('circunferencia_circulo', 1) == pytest.approx(6.283185307179586, rel=1e-9)

def test_calculadora_verificacao_triangulo():
    assert calculadora('verificacao_triangulo', 3, 4, 5) == True
    assert calculadora('verificacao_triangulo', 1, 1, 3) == False

# Operações de Análise Numérica e Financeira
def test_calculadora_verificacao_numeros_primos():
    assert calculadora('verificacao_numeros_primos', 7) == True
    assert calculadora('verificacao_numeros_primos', 4) == False

def test_calculadora_conversor_unidades():
    assert calculadora('conversor_unidades', 1, 'milhas_para_km') == 1.60934
    assert calculadora('conversor_unidades', 10, 'metros_para_pes') == 32.8084

def test_calculadora_juros_simples():
    assert calculadora('juros_simples', 1000, 0.05, 1) == 50
    assert calculadora('juros_simples', 2000, 0.03, 2) == 120

def test_calculadora_juros_compostos():
    assert calculadora('juros_compostos', 1000, 0.05, 2) == pytest.approx(1102.5, rel=1e-9)
    assert calculadora('juros_compostos', 1500, 0.03, 3) == pytest.approx(1642.5, rel=1e-9)

def test_calculadora_soma_numeros_naturais():
    assert calculadora('soma_numeros_naturais', 5) == 15
    assert calculadora('soma_numeros_naturais', 10) == 55

def test_calculadora_fatorial():
    assert calculadora('fatorial', 5) == 120
    assert calculadora('fatorial', 0) == 1

def test_calculadora_permutacao():
    assert calculadora('permutacao', 5, 3) == 60
    assert calculadora('permutacao', 7, 4) == 840

def test_calculadora_combinacao():
    assert calculadora('combinacao', 5, 3) == 10
    assert calculadora('combinacao', 7, 4) == 35

def test_calculadora_conversor_bases():
    assert calculadora('conversor_bases', 10, 'decimal_para_binario') == '1010'
    assert calculadora('conversor_bases', 255, 'decimal_para_hexadecimal') == 'FF'

def test_calculadora_media():
    assert calculadora('media', 1, 2, 3, 4, 5) == 3
    assert calculadora('media', 10, 20, 30) == 20

def test_calculadora_desvio_padrao():
    assert calculadora('desvio_padrao', 1, 2, 3, 4, 5) == pytest.approx(1.4142135623730951, rel=1e-9)
    assert calculadora('desvio_padrao', 10, 20, 30) == pytest.approx(10.0, rel=1e-9)

def test_calcular_tangente_reta():
    assert calculadora('tangente_reta', 1, -1, 1) == -1
    assert calculadora('tangente_reta', 2, 3, -4) == -2.3333333333333335



```
